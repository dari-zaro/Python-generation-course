```python
#Задача 1

with open('C:\\Users\\THUNDEROBOT\\Downloads\\dataset_3363_2.txt') as t:
    s = t.readline().strip()
def rep(symbol, iter): return symbol*int(iter)
i = 0
while i < len(s):
    iter = ''
    if s[i].isalpha():
        symbol = s[i]
        i += 1
        while s[i].isdigit():
            iter += s[i]
            if i == len(s)-1: break
            i += 1
        print(rep(symbol, iter),end='')
```

```python
#Задача 2

words = open('C:\\Users\\THUNDEROBOT\\Downloads\\dataset_3363_3.txt').read().lower().split() 
dict_ = {} 
max_ = 0
for w in words:
    qty = dict_.get(w, 0) + 1
    dict_[w] = qty
    max_ = max(max_, qty)
max_items = sorted([item for item in dict_.items() if item[1] == max_])
print( max_items[0][0], max_items[0][1] )
```

```python
#Задача 3

with open('C:\\Users\\THUNDEROBOT\\Downloads\\dataset_3363_4.txt') as f:
    strings = [s.rstrip() for s in f.readlines()]
otsenki = [s.split(';')[1:] for s in strings]
for x in otsenki:
    print(sum(map(int, x))/len(x))
sr_mat = sum([int(x[0]) for x in otsenki]) / len(otsenki)
sr_fiz = sum([int(x[1]) for x in otsenki]) / len(otsenki)
sr_rus = sum([int(x[2]) for x in otsenki]) / len(otsenki)
print(sr_mat, sr_fiz, sr_rus)
```
